Notion de **constructeur**

```python
from tutor import tutor

class Exemple:
    def __init__(self):
        self.a = 10
        self.b = 20

charlie = Exemple()
charlie.a += 50
delta = Exemple()
delta.b += 100

tutor()
```

en python, le **constructeur** d'une classe : 
* est la méthode `__init__`
* qui est exécutée lors de la construction de chaque objet de la classe (ou instance de classe)
* et qui a pour argument `self` qui désigne l'objet lui-même qui sera construit.

#### autres arguments du constructeur

```python
from tutor import tutor

class Exemple:
    def __init__(self, x, y):
        self.a = x
        self.b = y

charlie = Exemple(10,20)
charlie.a += 50
delta = Exemple(30,40)
delta.b += 100

tutor()
```
