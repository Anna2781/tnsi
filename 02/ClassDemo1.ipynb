## notion d'attribut

```python
class Test:
    attr = 5
    
    def f(x):
        return (x.attr)
```

```python
a = Test()
print(a.attr)
```

```python
a.attr = 10
print(a.attr)
```

```python
b = Test()
print(b.attr)
```

```python
b.attr += 1
print(b.attr)
```

A RETENIR :  
* une classe permet de définir des **attributs** pour chacun de leurs objets. 
* on accède à la valeur d'un attribut avec la **notation pointée** : `nom_objet.nom_attribut`

## notion de méthode

```python
a.f()
```

```python
b.f()
```
