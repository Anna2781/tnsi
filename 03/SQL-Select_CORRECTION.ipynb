### Bases de données : REQUÊTES d'interrogation

```sql
CREATE TABLE x(a  INT,
               b  INT
);
CREATE TABLE y(c  INT,
               d  INT
);
CREATE TABLE z(a  INT,
               c  INT,
               e  INT
);

INSERT INTO x VALUES (1,1), (2,2), (3,2), (4,2), (5,1), (6,9), (7,1);
INSERT INTO y VALUES (9,9), (10,10), (11,9), (12,20), (13,30), (14,9), (15,1), (16,10), (17,10);
INSERT INTO z VALUES (1,11,30), (2,14,9),(5,15,1), (7,17,3), (1,10,50), (2,9,8), (2,15,15), (3,17,19), (4,16,12), (5,10,20), (2,11,30), (7,14,9), (7,9,12);
```

```sql
SELECT * FROM x WHERE b = 1;
```

```sql
SELECT * FROM z
```

```sql
SELECT  DISTINCT  e FROM z WHERE e>10 AND e<50;
```

```sql
SELECT * FROM y WHERE c%2=0 ORDER BY d ASC;
```

```sql
SELECT * FROM x JOIN z ON x.a=z.a  WHERE z.e<10
```

```sql
SELECT x.a, x.b FROM x JOIN z ON x.a=z.a  WHERE z.e<10
```

```sql
SELECT * FROM x JOIN z ON x.a=z.a JOIN y ON y.c=z.c;
```

```sql
SELECT DISTINCT x.b, y.d FROM x JOIN z ON x.a=z.a JOIN y ON y.c=z.c;
```
