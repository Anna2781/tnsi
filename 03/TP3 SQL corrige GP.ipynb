#### 2.Effectuer pour vous-même un emprunt, pour le livre que vous choisissez, et pour une durée de deux semaines.

```sql
INSERT INTO emprunt(code_barre, isbn, retour) 
VALUES (978942526703422, '978-0201914658', '2022-10-24');
```

la requête n'affiche rien.  
On vérifie l'exécution correcte avec la requête suivante :

```sql
SELECT usager.nom, usager.prenom, usager.code_barre, livre.titre, livre.isbn, emprunt.retour
FROM emprunt 
JOIN usager ON emprunt.code_barre = usager.code_barre 
JOIN livre ON emprunt.isbn = livre.isbn 
WHERE usager.nom = 'PIED';
```

#### 3.Prolonger cet emprunt d'une semaine.

```sql
UPDATE emprunt 
SET retour='2022-10-31' 
WHERE code_barre=978942526703422;
```

la requête n'affiche rien.  
On vérifie l'exécution correcte avec la requête suivante :

```sql
SELECT usager.nom, usager.prenom, usager.code_barre, livre.titre, livre.isbn, emprunt.retour 
FROM emprunt 
JOIN usager ON emprunt.code_barre = usager.code_barre 
JOIN livre ON emprunt.isbn = livre.isbn 
WHERE usager.nom = 'PIED';
```

#### 4.Effectuer pour vous-même l'emprunt d'un livre rendu par un autre usager : il faudra donc effectuer deux requêtes.

#### i.la première pour "retourner le livre", rendu par l'autre usager.

```sql
DELETE FROM emprunt 
WHERE isbn='978-2864973270';
```

#### ii.la deuxième pour "emprunter le livre" à votre tour.

```sql
INSERT INTO emprunt(code_barre, isbn, retour)
VALUES (978942526703422, '978-2864973270', '2022-10-31');
```

la requête n'affiche rien.  
On vérifie l'exécution correcte avec la requête suivante :

```sql
SELECT usager.nom, usager.prenom, usager.code_barre, livre.titre, livre.isbn, emprunt.retour 
FROM emprunt 
JOIN usager ON emprunt.code_barre = usager.code_barre 
JOIN livre ON emprunt.isbn = livre.isbn 
WHERE usager.nom = 'PIED';
```
