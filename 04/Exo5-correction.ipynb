## Exercice 5 : correction

```
class Maillon:  
    def __init__(self, val, suiv=None):
        self.valeur = val
        self.suivant = suiv

    def __str__(self):
        if self is None:
            return ''
        if self.suivant is None:
            return str(self.valeur)
        return str(self.valeur) + '-' + str(self.suivant)
    
lst = Maillon(10,Maillon(20))
```

1) Ecrire l'instruction permettant d'insérer un maillon de valeur 15 entre ceux de valeurs 10 et 20.  
Une fois cette instruction exécutée, `print(lst)` doit donner `10-15-20`

#### réponse

```
lst.suivant = Maillon(15, lst.suivant)
print(lst)
```

2)  Ecrire les instructions permettant d'insérer un maillon de valeur 25  APRÈS ceux de valeurs 10 15 20.  
Une fois ces instructions exécutées, `print(lst)` doit donner `10-15-20-25`

#### réponse

```
lst.suivant.suivant.suivant = Maillon(25)
print(lst)
```

#### remarque : 
cette solution convient dans le cas où lst comporte 3 maillons... 
mais en général, cette stratégie n'est pas possible !

3) Ecrire le code d'une fonction `ajoute_a_la_fin` qui :

- prend deux paramètres:
  - `lst` de type Maillon
  - `v` 
- et ajoute un nouveau maillon de valeur `v` après le dernier maillon de la chaîne `lst`.


#### principe des solutions
L'idée est d'explorer les maillons de suivant en suivant... jusqu'à None


### a) avec une boucle while

```
# code à compléter
def ajoute_a_la_fin(lst,v):
    # on explore les maillons de suivant en suivant jusqu'à None
    while ...  is not None:
        lst = lst.suivant
    ...  = Maillon(v)
    
```

```
def ajoute_a_la_fin(lst,v):
    # on explore les maillons de suivant en suivant jusqu'à None
    while lst.suivant is not None:
        lst = lst.suivant
    lst.suivant = Maillon(v)
```

```
lst = Maillon(10, Maillon(20))
ajoute_a_la_fin(lst, 30)
```

### b) avec une fonction récursive

```
# code à compléter
def ajoute_a_la_fin(lst,v):
    # cas de base
    if ...  is None :
        lst.suivant = Maillon(v)
    # cas récursif
    else : 
        ajoute_a_la_fin( ... , ... )
```

```
def ajoute_a_la_fin(lst,v):
    # cas de base
    if lst.suivant is None :
        lst.suivant = Maillon(v)
    # cas récursif
    else : 
        ajoute_a_la_fin( lst.suivant , v)
```

```
lst = Maillon(10, Maillon(20))
ajoute_a_la_fin(lst, 30)
```
