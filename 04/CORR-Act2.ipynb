## Exercice 2

Écrire une fonction `longueur(lst)` qui utilise une boucle **while** pour renvoyer la longueur d'une liste, c'est à dire le nombre de maillons qu'elle contient. On convient que `None` représente une liste de longueur 0. 

```
# version itérative du calcul de la longueur

class Maillon:
    def __init__(self, val, suiv=None):
        self.valeur = val
        self.suivant = suiv

lst30 = Maillon(30, Maillon(20, Maillon(10)))

def longueur(lst):
    lg = 0
    while lst is not None :
        lst = lst.suivant
        lg += 1
    return lg

print(longueur(lst30))
    
```

### Exercice 2 (bis)
Ecrire une fonction total, qui renvoie la somme des valeurs de tous les maillons d'une chaîne

```
def total(lst):
    '''renvoie la somme des valeurs de tous les maillons de lst'''
    t = 0
    while lst is not None :
        t += lst.valeur
        lst = lst.suivant
    return t
```

```
lst = Maillon(30, Maillon(20, Maillon(10)))
```
