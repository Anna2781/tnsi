## Exercice 6 : correction

```
class Maillon:
    def __init__(self, val, suiv=None):
        self.valeur = val
        self.suivant = suiv

    def __str__(self):
        if self.suivant is None:
            return str(self.valeur)
        return str(self.valeur) + '-' + str(self.suivant)
```

Écrire une fonction occurrences(x, lst) qui renvoie le nombre d’occurrences d’une valeur x dans une liste chaînée lst.  
a) avec une fonction récursive  
b) avec une boucle while

## a) fonction récursive `occurrences`

```
def occurrences(x,lst):
    if lst is None:
        return 0
    elif lst.valeur == x:
        return 1 + occurrences(x, lst.suivant)
    else : 
        return occurrences(x, lst.suivant)
    
lst1 = Maillon(1, Maillon(2,Maillon(5, Maillon(1,None ))))
```

```
occurrences(1,lst1)
```

```
occurrences(2,lst1)
```

```
occurrences(0,lst1)
```

## b) fonction `occurrences` avec une boucle `while`

```
def occurrences(x,lst):
    cpt = 0
    while lst is not None:
        if lst.valeur == x :
            cpt += 1
        lst = lst.suivant
    return cpt 
```

```
lst1 = Maillon(1, Maillon(2,Maillon(5, Maillon(1,None ))))
assert occurrences(1,lst1) == 2
assert occurrences(2,lst1) == 1
assert occurrences(0,lst1) == 0
print('test OK')
```
