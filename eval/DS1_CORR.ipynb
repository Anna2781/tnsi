    res = []
    res = [] + ajouter ('a', produit('ab', 1))
    res = [] + ajouter ('b', produit('ab', 1))


Cela revient à exécuter : 


    res = []
    res = [] + ajouter ('a', ["a","b"])
    res = [] + ajouter ('b', ["a","b"])
    


ou encore : 

    res = []
    res = [] + ["aa", "ab"]
    res = [] + ["ba", "bb"]

finalement, on a donc res = []+["aa", "ab"]+["bb", "bb"]  
donc **`produit("ab", 2)` renvoie  ["aa", "ab", "ba", "bb"]**

```python
def produit(s, n):
    if n == 0:
        return [""]
    else : 
        res = []
        for i in range(len(s)):
            res = res + ajouter (s[i], produit(s, n-1))
        return res
produit("ab", 2)
```

1.a.  
Si tous les éléments du tableau sont positifs, la sous-séquence de valeur maximale s'obtient avec **toutes les valeurs du tableau**, et a pour valeur la somme de tous ces éléments. 

1.b.  
Si tous les éléments du tableau sont négatifs, la sous-séquence de valeur maximale (qui doit être non vide) est réduite à **un seul élément, dont la valeur est maximale** dans le tableau.

```python
# 2.a.
def somme_sous_sequence(lst,i,j):
    resu = 0
    for indice in range(i, j+1): # indice j inclus, donc indice j+1 exclu
        resu += lst[indice]
    return resu
```

```python
# 2.b.
def pgsp(lst):
    n = len(lst)
    somme_max = lst[0]
    for i in range(n):
        for j in range(i,n):
            s = somme_sous_sequence(lst, i, j)
            if s > somme_max :
                somme_max = s
    return somme_max
```

si le tableau `lst` a pour longueur 10, alors la comparaison `if s > somme_max :` est exécutée **55 fois**, puisque les deux boucles  `for i in range(10):` et `for j in range(i,10):` sont imbriquées : le nombre d'itération est 10+9+8+...+1 = 55.

```python
# 2.c. Amélioration du code pour renvoyer aussi les indices
def pgsp(lst):
    n = len(lst)
    somme_max = lst[0]
    i_optimal = 0
    j_optimal = 0
    for i in range(n):
        for j in range(i,n):
            s = somme_sous_sequence(lst, i, j)
            if s > somme_max :
                somme_max = s
                i_optimal = i
                j_optimal = j
    return (somme_max, i_optimal, j_optimal)
```

```python
# 3.b. 
def pgsp2(lst):
    sommes_max = [lst[0]]
    # i décrit tous les indices valides du tableau à partir de 1
    for i in range(1, len(lst)) : 
        if sommes_max[i-1] <= 0 :
            sommes_max.append(lst[i])
        else : 
            sommes_max.append(lst[i] + sommes_max[i-1])
    return max(sommes_max)
```
