# Exercice 1

```python
def correspond(mot, mot_a_trous):
    if len(mot) != len(mot_a_trous):
        return False
    for i in range(len(mot_a_trous)):
        if mot[i] != mot_a_trous[i]:
            if mot_a_trous[i] != '*':
                return False
    return True
```

```python
correspond('INFORMATIQUE', 'IN*O**ATI*UE')
```

```python
correspond('ABC', '**D')
```

```python
correspond('INFORMATIQUE', 'IN*O**ATI*WE')
```

# Exercice 2

```python
plan_a = {'A':'E', 'B':'F', 'C':'D', 'D':'C', 'E':'B', 'F':'A'}
plan_b = {'A':'C', 'B':'F', 'C':'E', 'D':'A', 'E':'B', 'F':'D'}
```

```python
def est_cyclique(plan):
    '''
    Prend en paramètre un dictionnaire `plan` correspondant 
    à un plan d'envoi de messages entre `N` personnes A, B, C, 
    D, E, F ...(avec N <= 26).
    Renvoie True si le plan d'envoi de messages est cyclique
    et False sinon. 
    '''
    personne = 'A'
    print(personne)
    N = len(plan)                          
    for i in range(N-1):
        if plan[personne] == 'A':
            return False
        else:
            personne = plan[personne]
            print(personne)
    return True
```

```python
est_cyclique(plan_a)
```

```python
est_cyclique(plan_b)
```
