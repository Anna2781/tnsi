# Exercice 1

```python
def conv_bin(n):
    lst_bin = []
    while n != 0 :
        b = n%2
        n = n //2
        lst_bin.append(b)
    lst_bin.reverse()
    return (lst_bin, len(lst_bin))
```

```python
conv_bin(254)
```

# Exercice 2

```python
def tri_bulles(T):
    n = len(T)
    for i in range(n-1,0,-1): #range(...,...,-1)
        for j in range(i):
            if T[j] > T[j+1]:
                temp = T[j]
                T[j] = T[j+1]
                T[j+1] = temp
    return T
```

```python
tri_bulles([10,50,30,40,20,10])
```
