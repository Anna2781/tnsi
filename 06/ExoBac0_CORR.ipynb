```python
class Maillon:
    def __init__(self, val, suiv=None):
        self.valeur = val
        self.suivant = suiv

    def __str__(self):
        if self.suivant is None:
            return str(self.valeur)
        return str(self.valeur) + '\n' + str(self.suivant)

class Pile:
    def __init__(self):
        self.haut = None

    def est_vide(self):
        return self.haut is None

    def empiler(self, val):
        self.haut = Maillon(val, self.haut)

    def __str__(self):
        return str(self.haut)


def creer_pile():
    return Pile()

def pile_est_vide(p):
    return p.est_vide()

def empiler(p,v):
    p.empiler(v)

def depiler(p):
    if p.est_vide():
        raise ValueError("impossible de d√©piler une pile vide")
    premier_maillon = p.haut
    p.haut = premier_maillon.suivant
    return premier_maillon.valeur
```

```python
def max_pile(P,i): #renvoie la position du max d'une pile P
    Q = creer_pile()
    max = 0
    for a in range(i):
        x = depiler(P)
        if max < x :
            imax = a + 1
            max = x
        empiler(Q,x)
    while not(pile_est_vide(Q)):
        empiler(P, depiler(Q))
    return imax

def retourner(P, i):
    Q = creer_pile()
    O = creer_pile()
    for i2 in range(i):
        empiler(Q, depiler(P))
    while not(pile_est_vide(Q)):
        empiler(O, depiler(Q))
    while not(pile_est_vide(O)):
        empiler(P, depiler(O))
    return P

def hauteur_pile(P):
    Q = creer_pile()
    n = 0
    while not(pile_est_vide(P)):
        n+=1
        x = depiler(P)
        empiler(Q,x)
    while not(pile_est_vide(Q)):
        empiler(P,depiler(Q))
    return n

def crepe_miam(P):
    h = hauteur_pile(P)
    while h != 0 :
        max_temp = max_pile(P, h)
        retourner(P, max_temp)
        retourner(P, h)
        h -= 1
    return print(P.haut)
```

```python
P = creer_pile()
empiler(P,8)
empiler(P,25)
empiler(P,2)
empiler(P,40)
```
