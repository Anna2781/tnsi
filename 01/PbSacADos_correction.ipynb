# correction : sac à dos récursif

```python
objlist = [('A', 3, 0.5), 
           ('B', 5,1), 
           ('C', 12,1), 
           ('D', 30,5), 
           ('E', 9,6), 
           ('F', 10,5), 
           ('G', 12,3)]

POIDS_MAX=10

def nom(objet):
    return objet[0]

def valeur(objet):
    return objet[1]

def masse(objet):
    return objet[2]

def affich(objlist):
    return [ v[0] for v in objlist]

def optimise(liste_obj_restant, sac, valeur_courante, masse_restante, prof):
    space = ' * '
    print(prof * space, f'optimise({affich(liste_obj_restant)},{affich(sac)}, {valeur_courante}, {masse_restante})')
    
    # cas de base 1 : on a épuisé la liste d'objets
    if len(liste_obj_restant)==0:
        print(prof * space, "sac vide")
        return valeur_courante, sac
    # cas récursif trivial : le premier objet de la liste est trop lourd 
    objet = liste_obj_restant[0]
    if masse(objet)>masse_restante : 
        print(prof * space, nom(objet), "trop lourd")
        return optimise(liste_obj_restant[1:],sac, valeur_courante, masse_restante, prof )
    else : # cas récursif nécessitant une comparaison
        # avec l'objet
        v1, s1 = optimise(liste_obj_restant[1:],sac+[objet], valeur_courante+valeur(objet), masse_restante-masse(objet), prof+1 )
        # sans l'objet
        v0, s0 = optimise(liste_obj_restant[1:],sac, valeur_courante, masse_restante , prof+1)
        if v0 > v1 : 
            return v0,s0
        else :
            return v1,s1

optimise(objlist, [], 0, POIDS_MAX, 0)
```
